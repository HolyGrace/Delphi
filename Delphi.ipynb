```python
import pandas as pd
import numpy as np
from ast import literal_eval
```

```python
df = pd.read_csv(".\Data\movies_dataset.csv")
```

```python
df.info()
```

```python
# fillna
df['belongs_to_collection'] = df['belongs_to_collection'].fillna('{}')
# covert to dicts
df['belongs_to_collection'] = df['belongs_to_collection'].apply(literal_eval)
```

```python

df_collection = pd.json_normalize(df['belongs_to_collection'])
df_collection.rename(columns={'id': 'collection_id','name':'collection_name', 'poster_path':'collection_poster_path', 'backdrop_path': 'collection_backdrop_path'},inplace=True)
df_collection.head()
```

```python
df_collection.info()
```

```python
df.drop(['belongs_to_collection'], axis=1, inplace=True)
df = df.join(df_collection)
```
